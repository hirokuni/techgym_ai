# (3) gensimでword2vecのモデル作成をして、「プログラミング」という単語をベクトル表現して表示する。プログラミングと類似している単語の上位５個を出力する

```python
from gensim.models import Word2Vec
from janome.tokenizer import Tokenizer
```

```python
text = Tokenizer()
```

```python
results = []
with open('./techgymAI.txt', encoding='Shift-JIS') as f:
    lines = f.readlines()
    for line in lines:
        text_c = t.tokenize(line, wakati=True)
        results.append(text_c)
```

```python
model = Word2Vec(results, min_count=1)
vector = model.wv['プログラミング']
print(vector)
```

```python
pro = model.wv.most_similar(positive=['プログラミング'], topn=5)
pro
```

```python
pro = model.wv.similar_by_vector('プログラミング')
pro
```

```python
pro = model.wv.similar_by_word('プログラミング')
pro
```
