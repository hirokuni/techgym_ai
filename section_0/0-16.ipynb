```python
import pandas as pd
import urllib
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import linear_model
```

```python
data = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data"
wine = pd.read_csv(data)

pd.set_option('display.max_rows', 5)
display(wine)

text = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.names"
txt = urllib.request.urlretrieve(text, './wine.names')
```

```python
with open('./wine.names') as f:
    for line in f:
        print(line)
```

```python
columns_name = ['class','Alcohol','Malic_acid','Ash',
                'Alcalinity_of_ash','Magnesium','Total_phenols',
                'Flavanoids','Nonflavanoid_phenols','Proanthocyanins',
                'Color_intensity','Hue','OD280_OD315','Proline']
wine.columns = columns_name
```

```python
REG = linear_model.LinearRegression()
```

```python
# display(wine['Alcohol'])
display(type(wine['Alcohol'])) # Series
#display(wine.loc[:,['Alcohol']])
display(type(wine.loc[:,['Alcohol']])) # DataFrame

X = wine.loc[:,['Alcohol']].values # èª¬æ˜Žå¤‰æ•°
# X = wine['Alcohol'].values # èª¬æ˜Žå¤‰æ•°
display(wine.loc[:,['Alcohol']].values)
display(wine['Alcohol'].values)

Y = wine['Color_intensity'].values # ç›®çš„å¤‰æ•°
#display(wine['Color_intensity'].values)
```

```python
REG.fit(X,Y)
```

```python
display('å›žå¸°ä¿‚æ•°: ', REG.coef_)
```

```python
display('åˆ‡ç‰‡: ', REG.intercept_)
```

```python
plt.xlabel('Alcohol')
plt.ylabel('Color intensity')
plt.scatter(X,Y)
plt.plot(X, REG.predict(X))
display("æ±ºå®šä¿‚æ•°: ", REG.score(X,Y))
```
