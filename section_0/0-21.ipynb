```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
data = './train.csv'
house_train = pd.read_csv(data)
```

```python
data = './test.csv'
house_test = pd.read_csv(data)
```

```python
house_train.isnull().sum()
house_test.isnull().sum()
```

```python
house_train.info()
```

```python
# データ準備
x_ = house_train['OverallQual'].values
x = house_train[['OverallQual']].values
y= house_train['SalePrice'].values

print("type house_train : {}, values : {}, values [[]]: {}".format(type(house_train), type(x_), type(x)))
print(type(house_train))
```

```python
print('説明変数 x: {}'.format(x))
print('目的変数 y: {}'.format(y))
```

```python
# 線形回帰
model = LinearRegression()
```

```python
# fitでモデル生成
model.fit(x,y)
```

```python
print('回帰係数: {}'.format(model.coef_))
print('回帰係数 type: {}'.format(type(model.coef_)))
print('切片: {}'.format(model.intercept_))
```

```python
# テストデータ、NaNがあるところは前の値で埋める
x_test = house_test[['OverallQual']].values
print('xtest: {}'.format(x_test))
```

```python
display(house_test.info())
```

```python
y_test_pred = model.predict(x_test)
```

```python
# df_testにSalePrice絡むを追加し、学習モデルから予測した結果をセットする
house_test['SalePrice'] = y_test_pred 
```

```python
house_test[['Id', 'SalePrice']].to_csv('submission.csv', index=False)
```
