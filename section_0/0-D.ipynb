```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy.random as random
```

```python
init_seed = 0
x_num = 100
y_num = 100
x_fig = 10
y_fig = 10
```

```python
random.seed(init_seed)
```

```python
x = np.random.randn(x_num)
y = np.random.randn(y_num) + np.sin(x)
```

```python
plt.figure(figsize=(x_fig, y_fig))
#plt.scatter(x,y)
plt.plot(x,y,'o')
x = np.linspace(-5,5,100)
plt.plot(x, np.sin(x))
plt.grid(True)
plt.title('Title')
plt.xlabel('X')
plt.ylabel('Y')
```
