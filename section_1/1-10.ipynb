# (1) ファイル読み込みして表示する

```python
lines = None
with open('./techgymAI.txt', encoding='Shift-JIS') as f:
    lines = f.readlines()
    print(lines)
#     for line in lines:
#         print(line)
```

# (2) Janomeを使用して形態素解析を行い「分かち書き形式」で表示する

```python
from janome.tokenizer import Tokenizer
```

```python
t = Tokenizer()
```

```python
word_list = []
for line in lines:
    token = t.tokenize(line, wakati=True)
    word_list.append(token)
```

```python
print(word_list)
```

# (3) gensimでword2vecのモデルを作成して、「プログラミング」という単語をベクトル表現して表示しよう。また、プログラミングと類似している単語の上位５個を出力する

```python
from gensim.models import Word2Vec
```

```python
model = Word2Vec(word_list, min_count=1)
```

```python
vector = model.wv['プログラミング']
```

```python
print(vector)
```

```python
display(model.wv.most_similar(positive=['プログラミング'], topn=5))
```

```python
display(model.wv.similar_by_word('プログラミング', topn=5))
```

```python
display(model.wv.similar_by_vector('プログラミング', topn=5))
```
