```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
random_states = [1, 5, 10, 15, 20, 25]
X = [None for _ in range(len(random_states))]
y = [None for _ in range(len(random_states))]
for index, i in enumerate(random_states) :
    X[index],y[index] = make_blobs(random_state = i)
    plt.scatter(X[index][:,0], X[index][:,1], label=str(i))
    plt.legend()
```

```python
sse_5 = []
sse_15 = []
for cluster_num in range(1,11):
    kmeans = KMeans(n_clusters=cluster_num, init='random')
    
    kmeans.fit(X[1])
    y_cluster_5 = kmeans.predict(X[1])
    sse_5.append(kmeans.inertia_) 
    
    kmeans.fit(X[3])
    y_cluster_15 = kmeans.predict(X[3])
    sse_15.append(kmeans.inertia_) 
sse_5
```

```python
plt.plot(range(10), sse_5, color='blue', label='random state = 5')
plt.plot(range(10), sse_15, color='red', label='random state=15')
plt.ylabel('SSE')
plt.xlabel('total cluster number')
plt.legend()
```
