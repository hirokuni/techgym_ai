```python
import pandas as pd
```

```python
data = "http://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data"
df = pd.read_csv(data, header=None)

pd.set_option("display.max_rows", 5)
df.columns =  ['class','Alcohol','Malic_acid','Ash',
                'Alcalinity_of_ash','Magnesium','Total_phenols',
                'Flavanoids','Nonflavanoid_phenols','Proanthocyanins',
                'Color_intensity','Hue','OD280_OD315','Proline']
corr = df.corr()
```

# loc（名前指定)  
ただし、indexに名前がつけられていない時は**[0, 1]**や**0:1**のように指定する

```python
display(df)
```

```python
# 全行、列指定(名)
display(df.loc[:, ['class','Alcohol']])
# 行範囲（番号ー名）、列指定(名)
display(df.loc[0:10, ['class','Alcohol']])  #display(df.loc[0:10, 0:2]) <- Error
# 行範囲、列範囲(名)
display(df.loc[0:2, 'class':'Ash']) # display(df.loc[0:2, 0:3])
```

```python
display(corr)
```

```python
# 行範囲 (名前)
display(corr.loc['class':"Malic_acid"])
# 行範囲（名前）、列指定(名前)
display(corr.loc['Alcohol':"Color_intensity", ['class','Malic_acid']]) # corr.loc[0:1 ['class','Malic_acid']] <- Error
# 行範囲（名前）、列範囲(名前)
display(corr.loc['Alcohol':"Color_intensity", 'class':'Malic_acid'])# corr.loc[0:1 ['class','Malic_acid']] <- Error
```

```python
# 行範囲（名前）、列範囲(名前)
corr.loc['Alcohol':"Color_intensity", 'class':'Malic_acid'] 
```

# iloc (番号指定）

```python
# 全行、列名
display(df.iloc[:])
```

```python
# 行範囲（番号）、列名
display(df.iloc[0:10, 0:2]) # display(df.loc[0:10,['class','Alcohol']])
#行指定
display(df.iloc[[0,2]])
#行範囲、列指定
display(df.iloc[0:2, [0,3]])
#行範囲、列範囲
display(df.iloc[0:2, 0:4])
```

```python
display(corr)
```

```python
display(corr.iloc[0:3])             # corr.iloc['class':"Malic_acid"] <- Error
display(corr.iloc[0:3,0:3])
```
